```python
# in this code file, we tried to find the best value of k in caes p equals to 4.
# in this file, we implement Thread in python3, whcih is uesd for multiprocessors running parallelly
# we run models with 8 parameters simultaneously, which requires at least 4 cores and 8 processors
# or you shall change the code based on the configuration of the computer
import numpy as np
from datetime import datetime
from dataloader import *
from knnmodel import *
from sklearn.model_selection import ShuffleSplit
import _thread
import threading
```

```python
# key function of running knn for mnist dataset
# implement 5-folder cross validation
# the variable "result" is the value of time running this Knn function. 
def RunKnnmnist(threadName,k,p,trainxdata, trainydata):
    ## running knn
    if exitFlag:
            threadName.exit()
    a = datetime.now()
    rs = ShuffleSplit(n_splits=5, test_size=.2, random_state=0)
    accuracy = 0
    for train_index, test_index in rs.split(trainxdata):
        train_x=trainxdata[train_index]
        train_y=trainydata[train_index]
        test_x=trainxdata[test_index]
        test_y=trainydata[test_index]
        
        test_num = test_x.shape[0]
        matchCount = 0
        for i in range(test_num):
            predict = kNNClassify(test_x[i], train_x, train_y, k,p)
            if predict == test_y[i]:
                matchCount += 1
        accuracy = accuracy + (float(matchCount) / test_num)
    accuracy = float(accuracy) / 5
    b = datetime.now()
    accuracy
    result = (b-a).seconds
    print(threadName,accuracy,result)
```

```python
# load data
# since the mnist dataset contains 60000 images for training data and 10000 for testing
# it's too great to calculate the 4-order distance
# we only use half of the trainning dataset which is 30000 images to do the corss validation. 
trainxdata, trainydata, testxdata, testydata = loadmnist()
xtrain=trainxdata[0:30000]
ytrain=trainydata[0:30000]
xtest=testxdata[0:30000]
ytest=testydata[0:30000]
exitFlag = 0
```

```python
# using diferent processors running different model simultaneously
# and show the result
class myThread (threading.Thread):
    def __init__(self, threadID, name,k, p,xdata,ydata):
        threading.Thread.__init__(self)
        self.threadID = threadID
        self.name=name
        self.xdata=xdata
        self.ydata=ydata
        self.k = k
        self.p = p
    def run(self):
        print ("strat：" + self.name)
        RunKnnmnist(self.name, self.k, self.p,self.xdata,self.ydata)
        print ("exit：" + self.name)
threadLock = threading.Lock()
        
thread1 = myThread(1, "k=1&p=4", 1, 4, xtrain, ytrain,)
thread2 = myThread(2, "k=2&p=4", 2, 4, xtrain, ytrain,)
thread3 = myThread(3, "k=3&p=4", 3, 4, xtrain, ytrain,)
thread4 = myThread(1, "k=4&p=4", 4, 4, xtrain, ytrain,)
thread5 = myThread(2, "k=5&p=4", 5, 4, xtrain, ytrain,)
thread6 = myThread(3, "k=6&p=4", 6, 4, xtrain, ytrain,)
thread7 = myThread(1, "k=7&p=4", 7, 4, xtrain, ytrain,)
thread8 = myThread(2, "k=8&p=4", 8, 4, xtrain, ytrain,)



# 开启新线程
thread1.start()
thread2.start()
thread3.start()
thread4.start()
thread5.start()
thread6.start()
thread7.start()
thread8.start()

thread1.join()
thread2.join()
thread3.join()
thread4.join()
thread5.join()
thread6.join()
thread7.join()
thread8.join()
```
